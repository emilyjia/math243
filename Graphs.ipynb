```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
nyt = np.loadtxt('nyt_k_file.txt')
fox = np.loadtxt('fox_k_file.txt')
ap=np.loadtxt('ap_k_file.txt')
breitbart = np.loadtxt('breitbart_k_file.txt')
wapo = np.loadtxt('wapo_k_file.txt')
sources = [nyt, wapo, ap, fox, breitbart]
source_name = ["nyt", "wapo", "ap", "fox", "breitbart"]
%matplotlib inline  
```

```python
for source in sources:
    source = [x for x in wapo if float(x) > 0]
```

```python
fig, ax = plt.subplots(3,2, figsize=(10,15))
hist, bins = np.histogram(nyt, bins=30,  range=(50, 200))
ax[0][0].bar(bins[:-1], hist.astype(np.float32) / hist.sum(), width=(bins[1]-bins[0]), color='grey')
ax[0][0].set_title("nyt yule-k distribution")
ax[0][0].set_ylim([0, 0.2])

hist, bins = np.histogram(wapo, bins=30, range=(50, 200))
ax[0][1].bar(bins[:-1], hist.astype(np.float32) / hist.sum(), width=(bins[1]-bins[0]), color='grey')
ax[0][1].set_title("wapo yule-k distribution")
ax[0][1].set_ylim([0, 0.2])

hist, bins = np.histogram(breitbart, bins=30, range=(50, 200))
ax[1][0].bar(bins[:-1], hist.astype(np.float32) / hist.sum(), width=(bins[1]-bins[0]), color='grey')
ax[1][0].set_title("breitbart yule-k distribution")
ax[1][0].set_ylim([0, 0.2])

hist, bins = np.histogram(fox, bins=30, range=(50, 200))
ax[1][1].bar(bins[:-1], hist.astype(np.float32) / hist.sum(), width=(bins[1]-bins[0]), color='grey')
ax[1][1].set_title("fox yule-k distribution")
ax[1][1].set_ylim([0, 0.2])

hist, bins = np.histogram(ap, bins=30, range=(50, 200))
ax[2][0].bar(bins[:-1], hist.astype(np.float32) / hist.sum(), width=(bins[1]-bins[0]), color='grey')
ax[2][0].set_title("ap yule-k distribution")
ax[2][0].set_ylim([0, 0.2])
ax[2][1].set_visible(False)

plt.savefig("yulek.png")
```
